```python
## 제2회 실기 작업형 제2유형 

X_test = pd.read_csv("C:/Users/MJ/Desktop/새 폴더 (2)/빅분기_기출문제_2회_code&data_수정본/X_test.csv",
                       encoding='utf-8-sig')
X_train = pd.read_csv("C:/Users/MJ/Desktop/새 폴더 (2)/빅분기_기출문제_2회_code&data_수정본/X_train.csv",
                       encoding='utf-8-sig')
y_train = pd.read_csv("C:/Users/MJ/Desktop/새 폴더 (2)/빅분기_기출문제_2회_code&data_수정본/y_train.csv",
                       encoding='utf-8-sig')
X_train.head(10)
```

```python
X_num = X_train[['Customer_care_calls', 'Customer_rating', 'Cost_of_the_Product', 'Prior_purchases', 'Discount_offered', 'Weight_in_gms']]
X_cat = X_train[['Warehouse_block', 'Mode_of_Shipment', 'Product_importance', 'Gender']]
X_cat = pd.get_dummies(X_cat)
```

```python
test_num = X_test[['Customer_care_calls', 'Customer_rating', 'Cost_of_the_Product', 'Prior_purchases', 'Discount_offered', 'Weight_in_gms']]
test_cat = X_test[['Warehouse_block', 'Mode_of_Shipment', 'Product_importance', 'Gender']] 
test_cat = pd.get_dummies(test_cat)
```

```python
X_cat, test_cat = X_cat.align(test_cat, join='inner', axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler_minmax = MinMaxScaler()
scaler_minmax.fit(X_num) 
X_scaled_minmax_train = scaler_minmax.transform(X_num)
X_scaled_minmax_test = scaler_minmax.transform(test_num)
```

```python
X_final = pd.concat([pd.DataFrame(X_scaled_minmax_train), X_cat], axis=1)
test_final = pd.concat([pd.DataFrame(X_scaled_minmax_test), test_cat], axis=1)
```

```python
y_train = y_train['Reached.on.Time_Y.N']
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_final, y_train)
```

```python
pred_test = model.predict_proba(test_final)
pred_test_prob = pd.DataFrame(pred_test[:, 0], columns=['predict_prob'])
final_predict = pd.concat([X_test['ID'], pred_test_prob], axis=1)
print(final_predict)
```

```python
## 제3회 실기 작업형 제2유형 
import pandas as pd 
train = pd.read_csv("C:/Users/MJ/Desktop/새 폴더 (2)/[3] 기출문제 3회_code&data/3rd_TravelInsurancePrediction_train.csv",
                    encoding='utf-8-sig')

test = pd.read_csv("C:/Users/MJ/Desktop/새 폴더 (2)/[3] 기출문제 3회_code&data/3rd_TravelInsurancePrediction_test.csv",
                    encoding='utf-8-sig')

X = train[train.columns[1:9]]
y = train[['TravelInsurance']]

X_num = X[['Age', 'AnnualIncome', 'FamilyMembers', 'ChronicDiseases']]
X_cat = X[['Employment Type', 'GraduateOrNot', 'FrequentFlyer', 'EverTravelledAbroad']]
X_cat=pd.get_dummies(X_cat)
```

```python
test_num = test[['Age', 'AnnualIncome', 'FamilyMembers', 'ChronicDiseases']]
test_cat = test[['Employment Type', 'GraduateOrNot', 'FrequentFlyer', 'EverTravelledAbroad']]
test_cat = pd.get_dummies(test_cat)
#test_num = 
```

```python
X_cat, test_cat = X_cat.align(test_cat, join='inner', axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
minmax.fit(X_num)
X_scaled=minmax.transform(X_num)
test_scaled=minmax.transform(test_num)

X_final = pd.concat([pd.DataFrame(X_scaled), X_cat], axis=1)
test_final = pd.concat([pd.DataFrame(test_scaled), test_cat], axis=1)

y = y['TravelInsurance']

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_final, y)

pred_test = model.predict_proba(test_final)
```

```python
pred_test_prob = pd.DataFrame(pred_test[:, 1], columns= ['predict_prob'])
```

```python
pred_test_prob
```

```python
final_predict = pd.concat([test['ID'], pred_test_prob], axis=1)
```
